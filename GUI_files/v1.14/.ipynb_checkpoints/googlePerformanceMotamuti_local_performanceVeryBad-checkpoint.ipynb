```python



import time
import PySimpleGUI as sg
import speech_recognition as sr
r = sr.Recognizer()
import csv

#for text to speech
from gtts import gTTS
from playsound import playsound
import os
import os.path
import PIL.Image
import io
import base64
import csv
import random


font1= ("Arial", 16)
test_words = {"Medium" : [], "Hard":[], "Easy":[]}
testList = [['']]
fontForTestWord = ("SiyamRupali", 20)
testIndex = 0
wrongMessage = "Please Try Again!"
rightMessage = "You Are Right!"
score = 10
word_index = {"Easy":0, "Medium":0, "Hard":0}
test = {"-word-":[], "-letter-":[], "-number-":[], "-sentence-":[]}
event1 = '-letter-'  # takes the values: -letter-, -sentence-, -letter-, -number-
level = 'Easy'
words = []
letters = []
numbers = []
sentences = []
current_position = 0
# lis = []
testWord = []
total_record_hit = 0
total_right_count = 0
percentage = 0
isGoogle = True
engineName = 'Google'
testFrequency = 2




#=======================Reading words, lettes, words and sentences from csv file====================================
with open(r'..\database\numbers\banglaNumbers_sorted.csv','r', encoding= 'utf-8')as file:
    filecontent=csv.reader(file)
    for row in filecontent:
        numbers.append(row)
with open(r'..\database\letters\banglaLetters_sorted.csv','r', encoding= 'utf-8')as file:
    filecontent=csv.reader(file)
    for row in filecontent:
        letters.append(row)
with open(r'..\database\words\banglaWords_sorted.csv','r', encoding= 'utf-8')as file:
    filecontent=csv.reader(file)
    for row in filecontent:
        words.append(row)
with open(r'..\database\sentences\BanglaSentences_sorted.csv','r', encoding= 'utf-8')as file:
    filecontent=csv.reader(file)
    for row in filecontent:
        sentences.append(row)

imageFileNames_wo_ext = []
imageFileNames_wo_dir = []
dir_path = '..\database\images'
for path in os.listdir(dir_path):
    # check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        directory = os.path.join(dir_path, path)
        name = os.path.splitext(path)[0]
        imageFileNames_wo_ext.append([name, directory] )
        imageFileNames_wo_dir.append(name)

#words
test["-word-"] = words
test["-letter-"] = letters
test["-number-"] = numbers
test["-sentence-"] = sentences


selected_chunk = test['-letter-']
learnedList = []


# engine_menu = [
#     ['&Settings',[['&Engine', ['&Google', '&Local']],['&Test Frequency', [['&One', '&Two', '&Three', '&Four', '&Five', '&Six']] ]] ]
# ]
engine_menu = [
    ['&Settings',[['&Engine', ['&Google', '&Local']]]]
]


sg.theme("Reddit")

#First Coloum
input_coloum =  [
    [sg.Menu(engine_menu)],
    [sg.Text("Choose a Category")],
    [sg.Button("Numbers", key = '-number-',  disabled = False)],
    [sg.Button("Letters", key = '-letter-',  disabled = False)],
    [sg.Button("Words", key = '-word-', disabled = False)],
    [sg.Button("Sentences", key = '-sentence-', disabled = False)],
    [sg.Text(size = (30, 1), key = '-VIEW_ERROR-',pad = ((0,0),(250,0)), auto_size_text= True, font = ('Arial', 5))],
    [sg.Text("Engine: "+engineName,font=('Arial', 8), key = '-ENGINE_TEXT-', pad = ((0,0),(10,0)))]
]

#Second Coloum
test_coloum = [

    [sg.Text("Pronounce it in Bengali: ", font = font1, visible = True, key = "-PRON-")],
    [sg.Text('', key = '-TEST_WORD-', font = fontForTestWord, visible = True, auto_size_text= True)],
    [sg.Text('', key = "-TEST_TRANS-", visible = True, auto_size_text= True)],
    [sg.Button('Record', key = "-REC-", visible = True)],
    [sg.Text(' ', key = '-OUTPUT-', visible = True, auto_size_text= True)],

  
    [sg.Button("Show Score", key = "-RES-", visible = False)],
    [sg.Text("Your have got " + str(percentage)+"% right!", key = "-SCORE-", visible = False)],
    [sg.Button('Next', key = '-NEXT-', disabled = True)]
    ]
teach_column = [
    [sg.Text("Let's Learn something new")],
    [sg.Text(size = (50,1), key = '-TEXT_TO_LISTEN-', font = ('SiyamRupali', 25), auto_size_text= True)],
    [sg.Text(size = (30,1), key = '-TEXT_TO_LISTEN_TRANS-', auto_size_text= True, pad = ((0,0),(10,10)))],
    [sg.Image(size=(100,100),key = '-SHOW_IMAGE-')],
    [sg.Button("Listen", key = "-LISTEN_AUDIO-")],
    [sg.Button("Previous", key = "-LISTEN_PREV-"), sg.Button("Next", key = "-LISTEN_NEXT-")]
]


layout = [
    [sg.Column(input_coloum, size = (200,500)),
     sg.VSeperator(),
     sg.Column(teach_column, visible = False, key = "-TEACH_COLUMN-"),
     sg.VSeperator(),
     sg.Column(test_coloum, visible = False, key = "-TEST_COLUMN-")

    ]


]

score_layout = [
    [sg.Text("Here is Your Score ", score)]
]
window1 = sg.Window("Learn Bengali in the Easy Way", layout,  size = (1000, 500), element_justification ='left')




def convert_to_bytes(file_or_bytes, resize=None): # using system: window['-SHOW_IMAGE-'].update(data=convert_to_bytes(os.getcwd()+words[wordIndex][1], resize=(200,200)))
    '''
    Will convert into bytes and optionally resize an image that is a file or a base64 bytes object.
    Turns into  PNG format in the process so that can be displayed by tkinter
    :param file_or_bytes: either a string filename or a bytes base64 image object
    :type file_or_bytes:  (Union[str, bytes])
    :param resize:  optional new size
    :type resize: (Tuple[int, int] or None)
    :return: (bytes) a byte-string object
    :rtype: (bytes)
    '''
    if isinstance(file_or_bytes, str):
        img = PIL.Image.open(file_or_bytes)
    else:
        try:
            img = PIL.Image.open(io.BytesIO(base64.b64decode(file_or_bytes)))
        except Exception as e:
            dataBytesIO = io.BytesIO(file_or_bytes)
            img = PIL.Image.open(dataBytesIO)

    cur_width, cur_height = img.size
    if resize:
        new_width, new_height = resize
        scale = min(new_height/cur_height, new_width/cur_width)
        img = img.resize((int(cur_width*scale), int(cur_height*scale)), PIL.Image.ANTIALIAS)
    with io.BytesIO() as bio:
        img.save(bio, format="PNG")
        del img
        return bio.getvalue()

def mode(event):
    if event == "-TEST-":
        window1["-TEACH_COLUMN-"].update(visible = False)
        window1['-TEST_COLUMN-'].update(visible = True)
    else:
        window1["-TEACH_COLUMN-"].update(visible = True)
        window1['-TEST_COLUMN-'].update(visible = False)


def initiate_test(learnedList):
    print('initiated test')
    global testWord
    testWord = random.choices(learnedList)

    print(testWord)
    mode('-TEST-')
    window1['-TEST_WORD-'].update(testWord[0][0])
    window1['-TEST_TRANS-'].update('In English>> ' + testWord[0][1])


def detectText(fromGoogle, word):

    if fromGoogle == word:
        return True
    else:
        return False

fromGoogle = ''

def record_and_recognize_new(testWord):
    print('from google')
    global score
    global fromGoogle
    global total_right_count
    isRight = False
    didSearch = False
    window1["-RES-"].update(visible = True)
    try:
        with sr.Microphone() as source:
            # read the audio data from the default microphone
            audio_data = r.record(source, duration = 3)
            print("Recognizing...")
            fromGoogle = r.recognize_google(audio_data, language = "bn-IN")
            didSearch = True
            print("Recognized Text:"+fromGoogle+"\\")
            print("database text:"+testWord[0] +"\\")

    except:

        window1['-OUTPUT-'].update(wrongMessage+" exeption", visible = True)
        score -= 1
        window1.refresh()

    if detectText(fromGoogle, testWord[0]):
        isRight = True
    elif  event1 == '-number-':
        if isinstance(int(testWord[1]), int) and detectText(fromGoogle, testWord[1]):
            print('256 test')
            isRight = True
        elif testWord[0] == '১' and fromGoogle in ['এক', '1']:
            isRight = True
        elif testWord[0] == '২' and fromGoogle in ['দুই', '2']:
            isRight = True
        elif testWord[0] == '৩' and fromGoogle in ['তিন', '3']:
            isRight = True
        elif testWord[0] == '৪' and fromGoogle in ['চার', '4']:
            isRight = True
        elif testWord[0] == '৫' and fromGoogle in ['পাঁচ', '5']:
            isRight = True
        elif testWord[0] == '৬' and fromGoogle in ['ছয়', '6']:
            isRight = True
        elif testWord[0] == '৭' and fromGoogle in ['সাত', '7']:
            isRight = True
        elif testWord[0] == '৮' and fromGoogle in ['আট', '8']:
            isRight = True
        elif testWord[0] == '৯' and fromGoogle in ['নয়', '9']:
            isRight = True
        elif testWord[0] == '১০' and fromGoogle in ['দশ', '10']:
            isRight = True
    elif event1 == '-sentence-':
        if detectText(fromGoogle, testWord[0][:-1]):
            isRight = True
            print('? detected from -sentence-') 
        elif detectText(fromGoogle, testWord[0][:-1].replace('কী', 'কি')):
            isRight = True
            print('? detected from replacing ki, kI')
    elif testWord[0][-1] == '?':
        print('? detected')        
        if detectText(fromGoogle, testWord[0][:-1]):
            isRight = True  
    elif testWord[0] == 'অ' and fromGoogle in ['ও', 'আহ', 'হ']:
        isRight = True
    elif testWord[0] in ['উ', 'ঊ'] and fromGoogle in ['ও', 'ই']:
        isRight = True
    elif testWord[0] in ['ই', 'ঈ'] and fromGoogle in ['ই']:
        isRight = True
    elif testWord[0] in ['ন', 'ণ'] and fromGoogle in ['ন', 'ণ']:
        isRight = True
    elif testWord[0] == 'ঋ' and fromGoogle in ['রি']:
        isRight = True
    elif testWord[0] == 'ঐ' and fromGoogle in ['অই', 'ওই']:
        isRight = True
    elif testWord[0] == 'ঙ' and fromGoogle in ['ইমো', 'ইমু']:
        isRight = True
        
    if isRight:
        window1['-OUTPUT-'].update(rightMessage, visible = True)
        score += 1
        total_right_count+=1
        window1['-VIEW_ERROR-'].update(visible = False)
        window1.refresh()
    if not isRight:
        window1['-OUTPUT-'].update(wrongMessage, visible = True)
        window1['-VIEW_ERROR-'].update("/"+testWord[0] + "/" + fromGoogle, visible=True)
        score -= 1
        window1.refresh()

import matlab.engine
eng = matlab.engine.start_matlab()
eng.recogMatlabInit()


def record_and_recognize_matlab(testWord):
    print('from matlab')
    global score
    global fromGoogle
    global total_right_count
    isRight = False
    didSearch = False
    window1["-RES-"].update(visible = True)
    try:
        print('trying :)')
        fromMatlab = eng.recogMatlab()
        print('recoged :3')
        fromGoogle = fromMatlab
        didSearch = True
        print("Recognized Text:"+fromGoogle+"\\")
        print("database text:"+testWord[0] +"\\")


    except:

        window1['-OUTPUT-'].update(wrongMessage+" exeption", visible = True)
        window1.refresh()

    if detectText(fromGoogle, testWord[0]):
        isRight = True
    elif  event1 == '-number-':
        if isinstance(int(testWord[1]), int) and detectText(fromGoogle, testWord[1]):
            print('256 test')
            isRight = True
        elif testWord[0] == '১' and fromGoogle in ['এক', '1']:
            isRight = True
        elif testWord[0] == '২' and fromGoogle in ['দুই', '2']:
            isRight = True
        elif testWord[0] == '৩' and fromGoogle in ['তিন', '3']:
            isRight = True
        elif testWord[0] == '৪' and fromGoogle in ['চার', '4']:
            isRight = True
        elif testWord[0] == '৫' and fromGoogle in ['পাঁচ', '5']:
            isRight = True
        elif testWord[0] == '৬' and fromGoogle in ['ছয়', '6']:
            isRight = True
        elif testWord[0] == '৭' and fromGoogle in ['সাত', '7']:
            isRight = True
        elif testWord[0] == '৮' and fromGoogle in ['আট', '8']:
            isRight = True
        elif testWord[0] == '৯' and fromGoogle in ['নয়', '9']:
            isRight = True
        elif testWord[0] == '১০' and fromGoogle in ['দশ', '10']:
            isRight = True
    elif event1 == '-sentence-':
        if detectText(fromGoogle, testWord[0][:-1]):
            isRight = True
        elif detectText(fromGoogle, testWord[0][:-1].replace('কী', 'কি')):
            isRight = True
    elif testWord[0][-1] == '?':
        print('? detected')
        if detectText(fromGoogle, testWord[0][:-1]):
            isRight = True        
    elif testWord[0] == 'অ' and fromGoogle in ['ও', 'আহ', 'হ']:
        isRight = True
    elif testWord[0] in ['উ', 'ঊ'] and fromGoogle in ['ও', 'ই']:
        isRight = True
    elif testWord[0] in ['ই', 'ঈ'] and fromGoogle in ['ই']:
        isRight = True
    elif testWord[0] == 'ঋ' and fromGoogle in ['রি']:
        isRight = True
    elif testWord[0] == 'ঐ' and fromGoogle in ['অই', 'ওই']:
        isRight = True
    elif testWord[0] == 'ঙ' and fromGoogle in ['ইমো', 'ইমু']:
        isRight = True
    if isRight:
        window1['-OUTPUT-'].update(rightMessage, visible = True)
        score += 1
        total_right_count+=1
        window1['-VIEW_ERROR-'].update(visible = False)
        window1.refresh()
    if not isRight:
        window1['-OUTPUT-'].update(wrongMessage, visible = True)
        window1['-VIEW_ERROR-'].update("/"+testWord[0] + "/" + fromGoogle, visible=True)
        score -= 1
        window1.refresh()

def play_audio(word_to_listen):
    tts = gTTS(word_to_listen, lang='bn')
    filename = 'temp.mp3'
    tts.save(filename)
    playsound(filename)

    path = os.getcwd()
    audio_path = path+'\\temp.mp3'

    if os.path.isfile(audio_path):
        os.remove(audio_path)


def pre_next_disable_check():
    if current_position >= len(selected_chunk) - 1:
        window1['-LISTEN_PREV-'].update(disabled=False)
        window1['-LISTEN_NEXT-'].update(disabled=True)
    elif current_position == 0:
        window1['-LISTEN_PREV-'].update(disabled=True)
        window1['-LISTEN_NEXT-'].update(disabled=False)
    elif current_position> 0 and current_position < len(selected_chunk)-1:
        window1['-LISTEN_PREV-'].update(disabled=False)
        window1['-LISTEN_NEXT-'].update(disabled=False)

def check_image_exists(wordName):
    global imageFileNames_wo_ext
    for rows in imageFileNames_wo_ext:
        if wordName == rows[0]:
            return True

    return False

def check_and_show_image():
    word_to_check = selected_chunk[current_position][0]
    if word_to_check[-1] == '?':
        word_to_check = word_to_check[:-1]
    print('word to check: '+word_to_check)
    if check_image_exists(word_to_check):
        img_index = imageFileNames_wo_dir.index(word_to_check)
        img_dir = imageFileNames_wo_ext[img_index][1]
        window1['-SHOW_IMAGE-'].update(data=convert_to_bytes(img_dir, resize=(100,100)), visible = True)
    if not check_image_exists(word_to_check):
        window1['-SHOW_IMAGE-'].update(visible = False)


def pre_next_function_teach(event):
    global current_position
    if event == '-LISTEN_NEXT-':
        current_position+=1
        pre_next_disable_check()
        check_and_show_image()

        window1['-TEXT_TO_LISTEN-'].update(selected_chunk[current_position][0])
        window1['-TEXT_TO_LISTEN_TRANS-'].update('In English>> ' + selected_chunk[current_position][1])
        window1.refresh()
    elif event == '-LISTEN_PREV-':
        current_position-=1
        pre_next_disable_check()
        check_and_show_image()
        window1['-TEXT_TO_LISTEN-'].update(selected_chunk[current_position][0])
        window1['-TEXT_TO_LISTEN_TRANS-'].update('In English>> ' + selected_chunk[current_position][1])



while True:
    event, values = window1.read()
    print(event)
    if event == sg.WIN_CLOSED:
        rows = []
        break
    print(values)
    if event in ['-number-', '-letter-', '-word-', '-sentence-']:
            event1 = event
            selected_chunk = test[event1]
            current_position = 0
            window1['-TEACH_COLUMN-'].update(visible = True)
            window1['-TEST_COLUMN-'].update(visible = False)  ######### added later
            window1['-OUTPUT-'].update(visible = False)
            check_and_show_image()
            window1['-TEXT_TO_LISTEN-'].update(selected_chunk[current_position][0],visible = True)
            window1['-TEXT_TO_LISTEN_TRANS-'].update('In English>> '+selected_chunk[current_position][1],visible = True)

    if event == '-LISTEN_AUDIO-':
        play_audio(selected_chunk[current_position][0])
        learnedList.append(selected_chunk[current_position])
    if event in ['-LISTEN_NEXT-', '-LISTEN_PREV-' ]:
        if len(learnedList)>= testFrequency:
            initiate_test(learnedList)
            learnedList = []
        pre_next_disable_check()
        pre_next_function_teach(event)

    if event == "-RES-":
        window1.refresh()
        if total_record_hit == 0:
            percentage = 0
        else:
            percentage = round(total_right_count/total_record_hit * 100, 2)
        window1["-SCORE-"].update("You have got " + str(percentage)+" % right!", visible = True)
    if event == '-NEXT-':
        mode('-TEACH-')
        window1['-SCORE-'].update(visible = False)
        window1['-VIEW_ERROR-'].update(visible = False)
        pre_next_disable_check()
        check_and_show_image()
        window1['-TEXT_TO_LISTEN-'].update(selected_chunk[current_position][0],visible = True)
        window1['-TEXT_TO_LISTEN_TRANS-'].update('In English>> '+selected_chunk[current_position][1],visible = True)
        window1['-OUTPUT-'].update(' ')
    if event == 'Google':
        isGoogle = True;
        engineName = 'Google'
        window1['-ENGINE_TEXT-'].update("Engine: "+engineName)
    if event == 'Local':
        isGoogle = False;
        engineName = 'Local'
        window1['-ENGINE_TEXT-'].update("Engine: "+engineName)
    if event == "-REC-":

        total_record_hit+=1
        if isGoogle:
            record_and_recognize_new(testWord[0])
#             print('dekho')
#             print(testWord[0])
        else:
            record_and_recognize_matlab(testWord[0])
#             print('dekho2')
#             print(testWord[0])
        window1['-NEXT-'].update(visible = True, disabled = False)

window1.close()



```
